```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
```

```python
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
```

```python
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS', inplace=True)
```

```python
vendas.info()
```

```python
media_preco = vendas['Preço'].mean()
vendas.fillna({'Preço': media_preco}, inplace=True)
#vendas['Preço'].fillna(media_preco, inplace=True) "Descontinuado"
```

```python
vendas.info()
```

```python
vendas[['Preço', 'Preço_com_frete']].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_corretas = vendas.query("Preço <= Preço_com_frete")
```

```python
vendas_corretas.describe()
```

```python
vendas_corretas.head()
```

```python
vendas_corretas.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
vendas_corretas.groupby('Nome_Departamento').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_corretas.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(), 2)
```

```python
vendas_corretas.groupby('Data').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
vendas_corretas['mes'] = vendas_corretas['Data'].dt.month
vendas_corretas.head()
```

```python
vendas_corretas.groupby('mes').idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
cliente[['idade', 'renda']].describe()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_corretas.merge(cliente, how='left', on='cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')[ 'renda'].agg('mean').sort_values(ascending=False).reset_index(), 2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')[ 'idade'].agg('mean').sort_values(ascending=False).reset_index(), 2)
agg_idade_bandeira
```

```python
fig = plt.figure(figsize=(7, 5))
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color='green')
plt.ylabel('Média de Idade')
plt.title('Média de Idade por Bandeira', loc='center', fontsize=18, fontweight='bold', color='black');
plt.show()
```

```python
fig = plt.figure(figsize=(7, 5))
plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color='green')
plt.ylabel('Média de Renda')
plt.title('Média de Renda por Canal de Venda', loc='center', fontsize=18, fontweight='bold', color='black');
plt.show()
```

```python
venda_por_data = vendas_corretas.groupby('Data').idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(15, 5))
plt.title('Quantidade de Vendas por Data', loc='center', fontsize=18, fontweight='bold', color='black')
plt.xlabel('Data da Venda')
plt.ylabel('Quantidade de Vendas')
x = venda_por_data['Data']
y = venda_por_data['idcompra']
plt.plot(x, y, color='green', linewidth='2')
plt.show()
```

```python
media_preco_por_departamento = round(vendas_corretas.groupby('Nome_Departamento')['Preço_com_frete'].agg('mean').sort_values(ascending=False).reset_index(), 2)
media_preco_por_departamento
```

```python
fig = plt.figure(figsize=(25, 5))
plt.title('Média de Renda por Canal de Venda', loc='center', fontsize=18, fontweight='bold', color='black');
plt.xlabel('Departamento')
plt.ylabel('Preço Médio')
x = media_preco_por_departamento['Nome_Departamento']
y = media_preco_por_departamento['Preço_com_frete']
plt.bar(x, y, color='green')
plt.show()
```
